```
# --- 1. Setup: Install Libraries & Import ---
# !pip install pandas matplotlib seaborn plotly ipywidgets

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from IPython.display import display, HTML # For displaying rich text in Colab
import io # Used for reading CSV data from string

# Configure plotting styles for a dark theme to match your dashboard aesthetic
plt.style.use('dark_background')
plt.rcParams.update({
    'text.color': 'lightgray',
    'axes.labelcolor': 'lightgray',
    'xtick.color': 'lightgray',
    'ytick.color': 'lightgray',
    'axes.edgecolor': 'lightgray',
    'figure.facecolor': '#1a1a1a', # Dark background for figure
    'axes.facecolor': '#1a1a1a',   # Dark background for plot area
    'grid.color': '#333333'        # Subtle grid lines
})
sns.set_palette('viridis') # A visually appealing and distinct color palette


# --- 2. Data Definition (Comprehensive Mock Data) ---
# This dataset is directly embedded, no external calls needed.
# It includes diverse examples to showcase various analytics and visuals.

print("--- Data Definition (Internal Mock Data) ---")

web_series_csv_data = """SeriesID,Title,Synopsis,Genres,StreamingPlatform,ReleaseYear,Creator,MainCast,IMDbRating,RottenTomatoesScore,PosterURL,TotalSeasons,EpisodeCount,Status,AwardsNominated,AwardsWon,AudienceEngagementScore
1,Stranger Things,"When a young boy vanishes, a small town uncovers a mystery involving secret experiments, terrifying supernatural forces, and one strange little girl.",Sci-Fi|Horror|Drama,Netflix,2016,Duffer Brothers,"Millie Bobby Brown, Finn Wolfhard, Winona Ryder",8.7,93,https://placehold.co/150x220/E50914/FFFFFF?text=Stranger+Things,4,34,Ongoing,200,80,85
2,The Witcher,"Geralt of Rivia, a mutated monster-hunter for hire, journeys toward his destiny in a turbulent world where people are often more wicked than beasts.",Fantasy|Action|Adventure,Netflix,2019,Lauren Schmidt Hissrich,"Henry Cavill, Anya Chalotra, Freya Allan",8.2,81,https://placehold.co/150x220/E50914/FFFFFF?text=The+Witcher,3,24,Ongoing,50,15,78
3,The Boys,"A group of vigilantes known informally as 'The Boys' set out to take down corrupt superheroes with no more than blue-collar grit and a willingness to fight dirty.",Action|Comedy|Crime,Amazon Prime Video,2019,Eric Kripke,"Karl Urban, Jack Quaid, Antony Starr",8.7,93,https://placehold.co/150x220/FF9900/FFFFFF?text=The+Boys,3,24,Ongoing,70,25,88
4,Squid Game,"Hundreds of cash-strapped contestants accept an invitation to compete in children's games for a tempting prize, but the deadly stakes soon become clear.",Action|Drama|Thriller,Netflix,2021,Hwang Dong-hyuk,"Lee Jung-jae, Park Hae-soo, Jung Ho-yeon",8.0,95,https://placehold.co/150x220/E50914/FFFFFF?text=Squid+Game,1,9,Ongoing,120,50,92
5,Money Heist,"An unusual group of robbers attempt to carry out the most perfect robbery in Spanish history - stealing 2.4 billion euros from the Royal Mint of Spain.",Action|Crime|Thriller,Netflix,2017,Álex Pina,"Úrsula Corberó, Álvaro Morte, Itziar Ituño",8.2,93,https://placehold.co/150x220/E50914/FFFFFF?text=Money+Heist,5,41,Completed,100,40,80
6,Ted Lasso,"American football coach Ted Lasso is hired to manage a British soccer team, Richmond F.C., despite having no experience.",Comedy|Sport,Apple TV+,2020,Bill Lawrence,"Jason Sudeikis, Hannah Waddingham, Brett Goldstein",8.8,92,https://placehold.co/150x220/000000/FFFFFF?text=Ted+Lasso,3,34,Ongoing,180,75,95
7,Bridgerton,"The eight close-knit siblings of the Bridgerton family look for love and happiness in London high society.",Drama|Romance,Netflix,2020,Chris Van Dusen,"Phoebe Dynevor, Regé-Jean Page, Nicola Coughlan",7.4,82,https://placehold.co/150x220/E50914/FFFFFF?text=Bridgerton,2,16,Ongoing,60,20,75
8,The Crown,"Follows the political rivalries and romance of Queen Elizabeth II's reign and the events that shaped the second half of the 20th century.",Drama|History,Netflix,2016,Peter Morgan,"Claire Foy, Olivia Colman, Imelda Staunton",8.7,86,https://placehold.co/150x220/E50914/FFFFFF?text=The+Crown,5,50,Ongoing,150,60,83
9,Loki,"The mercurial villain Loki resumes his role as the God of Mischief in a new series that takes place after the events of 'Avengers: Endgame'.",Action|Adventure|Fantasy,Disney+,2021,Michael Waldron,"Tom Hiddleston, Owen Wilson, Sophia Di Martino",8.2,92,https://placehold.co/150x220/003366/FFFFFF?text=Loki,2,12,Completed,40,10,87
10,The Mandalorian,"The travels of a lone bounty hunter in the outer reaches of the galaxy, far from the authority of the New Republic.",Action|Adventure|Sci-Fi,Disney+,2019,Jon Favreau,"Pedro Pascal, Carl Weathers, Giancarlo Esposito",8.7,93,https://placehold.co/150x220/003366/FFFFFF?text=Mandalorian,3,24,Ongoing,80,30,90
11,Panchayat,"A comedy-drama, which captures the journey of an engineering graduate who forgoes a cushy job for the role of a Gram Panchayat secretary in a remote village.",Comedy|Drama,Amazon Prime Video,2020,Deepak Kumar Mishra,"Jitendra Kumar, Raghubir Yadav, Neena Gupta",8.9,90,https://placehold.co/150x220/FF9900/FFFFFF?text=Panchayat,2,16,Ongoing,20,8,91
12,Mirzapur,"A shocking incident at a wedding procession ignites a war between two rival gangs in the city of Mirzapur, leading to a complex web of crime and power.",Action|Crime|Drama,Amazon Prime Video,2018,Karan Anshuman,"Pankaj Tripathi, Ali Fazal, Divyendu Sharma",8.5,85,https://placehold.co/150x220/FF9900/FFFFFF?text=Mirzapur,2,19,Ongoing,30,12,84
13,Scam 1992,"The story of Harshad Mehta, a stockbroker who single-handedly took the stock market to dizzying heights and then faced a dramatic downfall.",Biography|Crime|Drama,SonyLIV,2020,Hansal Mehta,"Pratik Gandhi, Shreya Dhanwanthary, Satish Kaushik",9.3,96,https://placehold.co/150x220/800080/FFFFFF?text=Scam+1992,1,10,Completed,110,45,94
14,Sacred Games,"A jaded cop is drawn into a web of betrayal and violence when he receives a cryptic warning from a gangster, leading him on a desperate hunt for answers.",Action|Crime|Drama,Netflix,2018,Vikramaditya Motwane,"Saif Ali Khan, Nawazuddin Siddiqui, Radhika Apte",8.5,91,https://placehold.co/150x220/E50914/FFFFFF?text=Sacred+Games,2,16,Completed,75,28,86
15,Kota Factory,"A look at the lives of students in Kota, a hub for coaching centers, as they prepare for competitive entrance exams.",Comedy|Drama,Netflix,2019,Raghav Subbu,"Jitendra Kumar, Mayur More, Ranjan Raj",9.0,94,https://placehold.co/150x220/E50914/FFFFFF?text=Kota+Factory,2,10,Ongoing,35,10,93
16,Arcane,"Set in the utopian region of Piltover and the oppressed underground of Zaun, the story follows the origins of two iconic League of Legends champions.",Animation|Action|Adventure,Netflix,2021,Christian Linke,"Hailee Steinfeld, Ella Purnell, Kevin Alejandro",9.0,100,https://placehold.co/150x220/E50914/FFFFFF?text=Arcane,1,9,Ongoing,130,55,96
17,Severance,"Mark Scout leads a team of office workers whose memories have been surgically divided between their work and personal lives.",Drama|Mystery|Sci-Fi,Apple TV+,2022,Dan Erickson,"Adam Scott, Zach Cherry, Britt Lower",8.7,97,https://placehold.co/150x220/000000/FFFFFF?text=Severance,1,9,Ongoing,90,35,89
18,Succession,"The lives of the Roy family, who own a media conglomerate, are explored as they contemplate what the future holds for them.",Drama,HBO Max,2018,Jesse Armstrong,"Brian Cox, Jeremy Strong, Sarah Snook",8.9,94,https://placehold.co/150x220/930000/FFFFFF?text=Succession,4,39,Completed,250,110,90
"""

df = pd.read_csv(io.StringIO(web_series_csv_data))
print("Raw Data Sample:")
display(df.head())


# --- 3. Data Cleaning & Preparation ---
print("\n--- Data Cleaning & Preparation ---")

# Convert ReleaseYear, TotalSeasons, EpisodeCount, Awards to numeric
numeric_cols = ['ReleaseYear', 'TotalSeasons', 'EpisodeCount', 'IMDbRating', 'RottenTomatoesScore',
                'AwardsNominated', 'AwardsWon', 'AudienceEngagementScore']
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Handle missing numerical values (e.g., fill with median)
for col in numeric_cols:
    df[col].fillna(df[col].median(), inplace=True)

# Create a 'CombinedRating' (normalized to 100-scale for consistency)
df['CombinedRating'] = (df['IMDbRating'] * 10 + df['RottenTomatoesScore']) / 2
df['CombinedRating'] = df['CombinedRating'].round(1) # Round to 1 decimal place

# Explode Genres for granular analysis (one row per genre per series)
df_genres_exploded = df.assign(Genres=df['Genres'].str.split('|')).explode('Genres')
df_genres_exploded['Genres'] = df_genres_exploded['Genres'].str.strip() # Clean whitespace

print("\nCleaned Data Info:")
df.info()
print("\nExploded Genres Data Sample:")
display(df_genres_exploded.head())


# --- 4. Advanced Analytics & Insights ---
print("\n--- Advanced Analytics & Insights ---")

# Insight 1: Top 5 Series by Combined Rating
top_rated_series = df.sort_values(by='CombinedRating', ascending=False).head(5)
print("\nTop 5 Series by Combined Rating (out of 100):")
display(top_rated_series[['Title', 'CombinedRating', 'IMDbRating', 'RottenTomatoesScore']])

# Insight 2: Average Combined Rating per Streaming Platform
avg_rating_platform = df.groupby('StreamingPlatform')['CombinedRating'].mean().sort_values(ascending=False).round(1)
print("\nAverage Combined Rating per Streaming Platform (out of 100):")
display(avg_rating_platform)

# Insight 3: Number of 'Ongoing' vs 'Completed' Series
status_counts = df['Status'].value_counts()
print("\nSeries Status Counts:")
display(status_counts)

# Insight 4: Most Awarded Creators (Top 3 by Awards Won)
creator_awards = df.groupby('Creator')['AwardsWon'].sum().sort_values(ascending=False).head(3)
print("\nTop 3 Creators by Awards Won:")
display(creator_awards)

# Insight 5: Correlation between Total Seasons and Combined Rating
correlation = df['TotalSeasons'].corr(df['CombinedRating'])
print(f"\nCorrelation between Total Seasons and Combined Rating: {correlation:.2f}")


# --- 5. Visualizations (Dashboard Layout Simulation) ---
print("\n--- Dashboard Generation Complete ---")
print("Displaying visualizations below. Please scroll to see all plots.")

# Use Plotly for interactive plots that simulate a dashboard page
# Plotly can render multiple figures if called sequentially or used with subplots.

# --- Visual 1: Interactive Catalog/Gallery (Main Display) ---
# Using Plotly for a richer, scrollable, and interactive catalog.
# This serves as the main "catalog" section of your dashboard.

print("\n--- Interactive Web Series Catalog ---")
fig_catalog = px.bar(df,
                     x='Title',
                     y='CombinedRating',
                     color='StreamingPlatform',
                     hover_data=['Synopsis', 'Genres', 'MainCast', 'ReleaseYear', 'TotalSeasons', 'Status', 'AwardsWon'],
                     labels={'CombinedRating': 'Combined Rating (out of 100)'},
                     title='Web Series Catalog: Rating Overview & Details on Hover',
                     height=600 # Adjust height to fit more bars initially
                    )
# Customizing hover template to include PosterURL (requires a bit more work in Plotly Express)
# For a full catalog *with images*, a custom HTML/CSS rendering is often best, but this gives interactivity.
fig_catalog.update_traces(marker_line_width=0, marker_line_color='black')
fig_catalog.update_layout(
    plot_bgcolor='#1a1a1a', paper_bgcolor='#1a1a1a', font_color='lightgray',
    title_font_color='lightgray',
    xaxis=dict(gridcolor='#333333', linecolor='#555555', tickangle=-45), # Rotate labels
    yaxis=dict(gridcolor='#333333', linecolor='#555555'),
    hoverlabel=dict(bgcolor="gray", font_size=12, font_family="Arial")
)
fig_catalog.show()

# You can also display a purely HTML-based image gallery:
html_image_catalog = "<h2>Visual Series Gallery</h2><div style='display: flex; flex-wrap: wrap; gap: 20px; justify-content: center;'>"
for index, row in df.iterrows():
    html_image_catalog += f"""
    <div style='border: 1px solid #333; border-radius: 8px; padding: 10px; width: 180px; background-color: #2a2a2a; color: lightgray; box-shadow: 3px 3px 8px rgba(0,0,0,0.7); text-align: center; overflow: hidden;'>
        <img src='{row['PosterURL']}' style='width: 100%; height: 220px; object-fit: cover; border-radius: 4px; margin-bottom: 8px;'>
        <h4 style='color: yellow; margin: 0 0 5px 0;'>{row['Title']} ({int(row['ReleaseYear'])})</h4>
        <p style='font-size: 0.8em; margin: 0 0 5px 0; color: #ADFF2F;'>Platform: {row['StreamingPlatform']}</p>
        <p style='font-size: 0.8em; margin: 0; color: #ADD8E6;'>Rating: {row['CombinedRating']}/100</p>
    </div>
    """
html_image_catalog += "</div>"
display(HTML(html_image_catalog))


# --- Remaining Visuals (Using Matplotlib/Seaborn for subplot layout) ---
# Create a figure for additional analytical plots
fig, axes = plt.subplots(2, 2, figsize=(20, 16), facecolor='#1a1a1a') # 2 rows, 2 columns

# --- Visual 2: Series Release Trend (Line Plot) ---
release_counts = df['ReleaseYear'].value_counts().sort_index()
sns.lineplot(x=release_counts.index, y=release_counts.values, marker='o', ax=axes[0, 0], color='cyan')
axes[0, 0].set_title('Series Release Trend Over Years', color='lightgray')
axes[0, 0].set_xlabel('Release Year', color='lightgray')
axes[0, 0].set_ylabel('Number of Series', color='lightgray')
axes[0, 0].grid(True, linestyle='--', alpha=0.5)

# --- Visual 3: Top 5 Genres by Series Count (Horizontal Bar Chart) ---
top_genres = df_genres_exploded['Genres'].value_counts().head(5)
sns.barplot(x=top_genres.values, y=top_genres.index, ax=axes[0, 1], palette='viridis')
axes[0, 1].set_title('Top 5 Genres by Series Count', color='lightgray')
axes[0, 1].set_xlabel('Number of Series', color='lightgray')
axes[0, 1].set_ylabel('Genre', color='lightgray')
for container in axes[0, 1].containers:
    axes[0, 1].bar_label(container, color='lightgray')


# --- Visual 4: Rating Distribution (Histogram with KDE) ---
sns.histplot(df['CombinedRating'], kde=True, ax=axes[1, 0], color='orange', bins=10)
axes[1, 0].set_title('Distribution of Combined Ratings (out of 100)', color='lightgray')
axes[1, 0].set_xlabel('Combined Rating', color='lightgray')
axes[1, 0].set_ylabel('Frequency', color='lightgray')
axes[1, 0].grid(True, linestyle='--', alpha=0.5)

# --- Visual 5: Awards Won by Platform (Stacked Bar Chart - as an example of a new chart type) ---
platform_awards = df.groupby('StreamingPlatform').agg(
    TotalAwardsWon=('AwardsWon', 'sum'),
    TotalAwardsNominated=('AwardsNominated', 'sum')
).reset_index()

# For a stacked bar, let's create a melted DataFrame
platform_awards_melted = platform_awards.melt(id_vars='StreamingPlatform',
                                            value_vars=['TotalAwardsWon', 'TotalAwardsNominated'],
                                            var_name='AwardType', value_name='Count')

sns.barplot(x='StreamingPlatform', y='Count', hue='AwardType', data=platform_awards_melted, ax=axes[1, 1], palette='pastel')
axes[1, 1].set_title('Awards Performance by Platform', color='lightgray')
axes[1, 1].set_xlabel('Streaming Platform', color='lightgray')
axes[1, 1].set_ylabel('Number of Awards', color='lightgray')
axes[1, 1].tick_params(axis='x', rotation=45) # Rotate x-axis labels if they overlap
axes[1, 1].legend(title='Award Type', title_fontsize='small', fontsize='small', facecolor='#2a2a2a', edgecolor='#333333')
axes[1, 1].grid(True, linestyle='--', alpha=0.5, axis='y') # Only Y-axis grid


plt.tight_layout() # Adjust layout to prevent overlapping
plt.show()

print("\n--- All Dashboard Visuals Generated ---")
```
