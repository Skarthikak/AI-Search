```
# --- 1. Setup: Install Libraries & Import ---
# !pip install pandas matplotlib seaborn plotly wordcloud

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from IPython.display import display, HTML
from wordcloud import WordCloud # For generating word clouds
import io # Used for reading CSV data from string
import numpy as np # For numerical operations

# Configure plotting styles for a dark theme, reminiscent of Nevermore
plt.style.use('dark_background')
plt.rcParams.update({
    'text.color': 'lightgray',
    'axes.labelcolor': 'lightgray',
    'xtick.color': 'lightgray',
    'ytick.color': 'lightgray',
    'axes.edgecolor': 'lightgray',
    'figure.facecolor': '#1a1a1a', # Dark background for figure
    'axes.facecolor': '#1a1a1a',   # Dark background for plot area
    'grid.color': '#333333'        # Subtle grid lines
})
sns.set_palette('viridis') # A visually appealing color palette

print("--- Wednesday Dashboard Report: A Night at Nevermore ---")


# --- 2. Data Definition (Comprehensive Mock Data for Wednesday Characters) ---
# This data simulates what you'd collect from various sources.
# PosterURL uses placehold.co for generic images to avoid real-time external image fetching.

print("\n--- Data Definition (Internal Mock Data for Wednesday Characters) ---")

wednesday_characters_csv_data = """CharacterID,CharacterName,ActorName,Synopsis,KeyAccomplishments,PersonalityTraits,IntelligenceScore,SocialAwkwardnessScore,CombatSkillsScore,MysterySolvingScore,LoyaltyScore,PosterURL
1,Wednesday Addams,Jenna Ortega,"A morbid and emotionally reserved teenager with psychic abilities, sent to Nevermore Academy where she uncovers a monstrous mystery.",Solved Hyde monster mystery; Prevented Nevermore's destruction; Mastered fencing; Played cello beautifully,"Sarcastic, Morbid, Independent, Observant, Goth",95,85,70,90,60,https://placehold.co/150x220/333333/93C5FD?text=Wednesday
2,Enid Sinclair,Emma Myers,"Wednesday's colorful and optimistic werewolf roommate who struggles with transforming, but proves fiercely loyal.",Befriended Wednesday; Transformed to save Wednesday; Helped solve mystery; Organized Rave'N Dance,"Optimistic, Colorful, Loyal, Social, Enthusiastic",70,20,50,60,95,https://placehold.co/150x220/8B0000/F0F8FF?text=Enid
3,Tyler Galpin,Hunter Doohan,"The charming and seemingly ordinary son of the sheriff, who develops a romantic interest in Wednesday, but hides a dark secret.",Worked at coffee shop; Befriended Wednesday; Revealed as Hyde; Followed orders of his master,"Charming, Secretive, Manipulative (as Hyde), Vulnerable",80,40,75,30,10,https://placehold.co/150x220/000080/FFFFFF?text=Tyler
4,Xavier Thorpe,Percy Hynes White,"A brooding Nevermore student with the power to bring his drawings to life, and a complex relationship with Wednesday.",Showed genuine concern for Wednesday; Helped investigate; Provided artistic clues,"Artistic, Brooding, Sensitive, Obsessive, Jealous",75,50,60,70,70,https://placehold.co/150x220/4B0082/FFFFFF?text=Xavier
5,Larissa Weems,Gwendoline Christie,"The elegant and stern principal of Nevermore Academy, a former roommate of Morticia Addams, and a shapeshifter.",Protected Nevermore's secrets; Attempted to guide Wednesday; Fought against evil plans; Sacrificed herself,"Authoritative, Protective, Secretive, Elegant, Observant",88,30,65,80,85,https://placehold.co/150x220/800000/FFFFFF?text=Weems
6,Eugene Otinger,Moosa Mostafa,"A quirky and kind Nevermore student, part of the 'Hummers' club, who becomes Wednesday's unlikely ally.",Helped Wednesday investigate; Provided crucial information; Survived Hyde attack,"Nervous, Loyal, Observant, Brave (when needed), Quirky",70,70,30,80,90,https://placehold.co/150x220/2F4F4F/FFFFFF?text=Eugene
7,Bianca Barclay,Joy Sunday,"A successful and popular Siren student at Nevermore, initially antagonistic towards Wednesday but later becomes an ally.",Excellent student; Leader of Sirens; Faced personal challenges; Competed with Wednesday,"Competitive, Confident, Popular, Talented, Proud",85,30,70,50,50,https://placehold.co/150x220/006400/FFFFFF?text=Bianca
8,Marilyn Thornhill,Christina Ricci,"A seemingly sweet and unassuming Nevermore dorm mom and botany teacher, with a hidden connection to the Hyde monster.",Provided guidance to students; Cultivated plants; Revealed as main antagonist; Manipulated Tyler,"Duplicitous, Manipulative, Deceptive, Intelligent, Obsessive",90,20,50,90,0,https://placehold.co/150x220/800000/FFFFFF?text=Thornhill
"""

df_wednesday = pd.read_csv(io.StringIO(wednesday_characters_csv_data))
print("Raw Data Sample:")
display(df_wednesday.head())


# --- 3. Data Cleaning & Preparation ---
print("\n--- Data Cleaning & Preparation ---")

# Convert score columns to numeric
score_cols = ['IntelligenceScore', 'SocialAwkwardnessScore', 'CombatSkillsScore', 'MysterySolvingScore', 'LoyaltyScore']
for col in score_cols:
    df_wednesday[col] = pd.to_numeric(df_wednesday[col], errors='coerce')
    df_wednesday[col].fillna(df_wednesday[col].median(), inplace=True) # Fill NaNs if any

# Explode PersonalityTraits for analysis (if needed for word clouds or trait counts)
df_traits_exploded = df_wednesday.assign(PersonalityTrait=df_wednesday['PersonalityTraits'].str.split(',')).explode('PersonalityTrait')
df_traits_exploded['PersonalityTrait'] = df_traits_exploded['PersonalityTrait'].str.strip()

print("\nCleaned Data Info:")
df_wednesday.info()
print("\nExploded Traits Data Sample:")
display(df_traits_exploded.head())


# --- 4. Advanced Analytics & Insights ---
print("\n--- Advanced Analytics & Insights ---")

# Insight 1: Overall Average Scores
print("\nOverall Average Character Scores:")
display(df_wednesday[score_cols].mean().round(1))

# Insight 2: Character with Highest Intelligence
highest_intel = df_wednesday.loc[df_wednesday['IntelligenceScore'].idxmax()]
print(f"\nCharacter with Highest Intelligence: {highest_intel['CharacterName']} ({highest_intel['IntelligenceScore']})")

# Insight 3: Most Common Personality Traits (Top 5)
top_traits = df_traits_exploded['PersonalityTrait'].value_counts().head(5)
print("\nTop 5 Most Common Personality Traits:")
display(top_traits)


# --- 5. Visualizations (Dashboard Layout Simulation) ---
print("\n--- Dashboard Visualizations ---")
print("Displaying visualizations below. Please scroll to see all plots.")

# --- Act I: The Gathering of Outcasts ---
display(HTML("<h3>Act I: The Gathering of Outcasts â€“ Meet the Nevermore Ensembles</h3><p style='color: #a0a0a0;'>Journey into the shadowy halls of Nevermore Academy and encounter its eccentric inhabitants. Our first scene introduces the peculiar students and staff who weave the intricate web of this supernatural world.</p>"))

# --- Visual 1: Character Catalog/Gallery (Main Display) ---
# Displaying character posters and basic info as a scrollable HTML gallery
print("\n--- Character Gallery ---")
html_character_gallery = "<h4>Roll Call: Faces of Nevermore</h4><div style='display: flex; flex-wrap: wrap; gap: 20px; justify-content: center;'>"
for index, row in df_wednesday.iterrows():
    html_character_gallery += f"""
    <div style='border: 1px solid #444; border-radius: 10px; padding: 12px; width: 200px; background-color: #2a2a2a; color: lightgray; box-shadow: 4px 4px 10px rgba(0,0,0,0.6); text-align: center; overflow: hidden;'>
        <img src='{row['PosterURL']}' style='width: 100%; height: 250px; object-fit: cover; border-radius: 8px; margin-bottom: 10px; border: 2px solid #555;'>
        <h4 style='color: #00FFFF; margin: 0 0 5px 0;'>{row['CharacterName']}</h4>
        <p style='font-size: 0.85em; margin: 0 0 5px 0; color: #ADD8E6;'>Actor: {row['ActorName']}</p>
        <p style='font-size: 0.75em; margin: 0; color: #D3D3D3;'>Scores: Int {row['IntelligenceScore']}, Soc {row['SocialAwkwardnessScore']}</p>
    </div>
    """
html_character_gallery += "</div>"
display(HTML(html_character_gallery))

# --- Act II: Whispers of the Past, Shadows of the Present ---
display(HTML("<h3 style='margin-top: 40px;'>Act II: Whispers of the Past, Shadows of the Present</h3><p style='color: #a0a0a0;'>As mysteries deepen, we uncover the essence of our protagonists and the threads of their influence. Who are they, beneath the surface?</p>"))

fig, axes = plt.subplots(2, 2, figsize=(20, 16), facecolor='#1a1a1a') # 2 rows, 2 columns

# --- Visual 2: Wednesday vs. Enid: Score Comparison (Bar Chart) ---
# This serves as a key character comparison, central to the series' dynamic.
wednesday_scores = df_wednesday[df_wednesday['CharacterName'] == 'Wednesday Addams'][score_cols].iloc[0]
enid_scores = df_wednesday[df_wednesday['CharacterName'] == 'Enid Sinclair'][score_cols].iloc[0]

comparison_df = pd.DataFrame({
    'Score Category': score_cols,
    'Wednesday': wednesday_scores.values,
    'Enid': enid_scores.values
})

comparison_melted = comparison_df.melt(id_vars='Score Category', var_name='Character', value_name='Score')

sns.barplot(x='Score Category', y='Score', hue='Character', data=comparison_melted, ax=axes[0, 0], palette='coolwarm')
axes[0, 0].set_title('Wednesday vs. Enid: A Clash of Personalities', color='lightgray')
axes[0, 0].set_xlabel('Score Category', color='lightgray')
axes[0, 0].set_ylabel('Score (out of 100)', color='lightgray')
axes[0, 0].tick_params(axis='x', rotation=45)
axes[0, 0].legend(title='Character', title_fontsize='small', fontsize='small', facecolor='#2a2a2a', edgecolor='#333333')
axes[0, 0].grid(True, linestyle='--', alpha=0.5, axis='y')


# --- Visual 3: Top Personality Traits Among Characters (Horizontal Bar Chart) ---
# Reveals common attributes that define the student body.
sns.barplot(x=top_traits.values, y=top_traits.index, ax=axes[0, 1], palette='flare')
axes[0, 1].set_title('Defining Traits: The Fabric of Nevermore', color='lightgray')
axes[0, 1].set_xlabel('Number of Characters Possessing Trait', color='lightgray')
axes[0, 1].set_ylabel('Trait', color='lightgray')
for container in axes[0, 1].containers:
    axes[0, 1].bar_label(container, color='lightgray')


# --- Act III: The Web Unravels, The Monster Awakens ---
display(HTML("<h3 style='margin-top: 40px;'>Act III: The Web Unravels, The Monster Awakens</h3><p style='color: #a0a0a0;'>The truth is rarely pretty. As secrets come to light, we analyze the impact of key actions and the hidden depths of their abilities.</p>"))

# --- Visual 4: Key Accomplishments Word Cloud ---
# Highlights the most significant events driven by character actions.
all_accomplishments = ' '.join(df_wednesday['KeyAccomplishments'].dropna().tolist())

stopwords_set = set([
    'a', 'an', 'the', 'is', 'in', 'of', 'to', 'and', 'with', 'from', 'for', 'on', 'who', 'his', 'her',
    'has', 'have', 'he', 'she', 'it', 'its', 'not', 'no', 'but', 'by', 'as', 'at', 'or', 'was', 'were',
    'are', 'been', 'had', 'hadn', 'hasn', 'haven', 'isn', 'wasn', 'weren', 'can', 'could', 'would',
    'should', 'may', 'might', 'must', 'my', 'me', 'our', 'us', 'we', 'they', 'them', 'their', 'this',
    'that', 'these', 'those', 'you', 'your', 'be', 'do', 'does', 'did', 'don', 'will', 'very', 'just',
    'about', 'all', 'any', 'most', 'some', 'such', 'only', 'own', 'too', 'up', 'down', 'out', 'in',
    'off', 'on', 'over', 'under', 'through', 'between', 'among', 'before', 'after', 'above', 'below',
    'to', 'from', 'with', 'without', 'into', 'onto', 'throughout', 'towards', 'upon', 'where', 'when',
    'why', 'how', 'what', 'which', 'whom', 'whose', 'myself', 'yourself', 'himself', 'herself', 'itself',
    'themselves', 'ourself', 'ourselves', 'myself', 'yourself', 'himself', 'herself', 'itself', 'themselves',
    'one', 'two', 'three', 'four', 'five', 'six', 'seven', 'eight', 'nine', 'ten', 'etc',
    # Specific to Wednesday's data
    'mystery', 'solved', 'helped', 'prevented', 'mastered', 'played', 'organized', 'uncovered',
    'befriended', 'transformed', 'save', 'revealed', 'followed', 'orders', 'master', 'genuine',
    'concern', 'provided', 'artistic', 'clues', 'protected', 'secrets', 'attempted', 'guide', 'fought',
    'sacrificed', 'herself', 'cultivated', 'plants', 'manipulated', 'survived', 'attack', 'leader',
    'sirens', 'faced', 'personal', 'challenges', 'competed'
])

wordcloud = WordCloud(width=800, height=400, background_color='#1a1a1a', colormap='winter',
                      stopwords=stopwords_set, # Use the correctly defined set
                      min_font_size=10).generate(all_accomplishments)

plt.figure(figsize=(12, 6), facecolor='#1a1a1a')
plt.imshow(wordcloud, interpolation='bilinear')
plt.title('Key Accomplishments: Echoes of Their Deeds', color='lightgray')
plt.axis('off')
plt.show()

# Continue with subplot figure created earlier
# --- Visual 5: Character Accomplishment Count (Horizontal Bar Chart) ---
# Shows which characters drove the most significant plot points.
accomplishment_counts = df_wednesday.assign(AccomplishmentCount=df_wednesday['KeyAccomplishments'].apply(lambda x: len(str(x).split(';')))).sort_values('AccomplishmentCount', ascending=False)

sns.barplot(x='AccomplishmentCount', y='CharacterName', data=accomplishment_counts, ax=axes[1, 0], palette='rocket')
axes[1, 0].set_title('Architects of Chaos & Order: Accomplishment Count', color='lightgray')
axes[1, 0].set_xlabel('Count of Accomplishments', color='lightgray')
axes[1, 0].set_ylabel('Character', color='lightgray')
for container in axes[1, 0].containers:
    axes[1, 0].bar_label(container, color='lightgray')


# --- Act IV: The Final Confrontation & Its Aftermath ---
display(HTML("<h3 style='margin-top: 40px;'>Act IV: The Final Confrontation & Its Aftermath</h3><p style='color: #a0a0a0;'>When the dust settles, what remains of their unique abilities and the overall balance of power? A look at the landscape of skills.</p>"))

# --- Visual 6: Intelligence vs. Mystery Solving (Interactive Scatter Plot) ---
# Visualizes the blend of intellect and investigative prowess.
fig_scatter = px.scatter(df_wednesday,
                         x='IntelligenceScore',
                         y='MysterySolvingScore',
                         size='LoyaltyScore', # Size points by Loyalty
                         color='CharacterName', # Color points by Character
                         hover_name='CharacterName',
                         title='Minds at Play: Intelligence vs. Mystery Solving',
                         labels={'IntelligenceScore': 'Intelligence Score', 'MysterySolvingScore': 'Mystery Solving Score'},
                         color_discrete_sequence=px.colors.qualitative.Pastel # Use a nice pastel palette
                        )
fig_scatter.update_layout(
    plot_bgcolor='#1a1a1a',
    paper_bgcolor='#1a1a1a',
    font_color='lightgray',
    title_font_color='lightgray',
    xaxis=dict(gridcolor='#333333', linecolor='#555555'),
    yaxis=dict(gridcolor='#333333', linecolor='#555555')
)
fig_scatter.show()

# --- Visual 7: Score Distribution (Box Plot for each score type) ---
# Provides an overall summary of score ranges and outliers.
df_scores_melted = df_wednesday.melt(id_vars=['CharacterName'], value_vars=score_cols, var_name='ScoreType', value_name='Score')

sns.boxplot(x='ScoreType', y='Score', data=df_scores_melted, ax=axes[1, 1], palette='mako')
sns.stripplot(x='ScoreType', y='Score', data=df_scores_melted, color='white', jitter=0.2, size=5, ax=axes[1, 1]) # Add individual points
axes[1, 1].set_title('The Spectrum of Abilities: Distribution of Character Scores', color='lightgray')
axes[1, 1].set_xlabel('Score Type', color='lightgray')
axes[1, 1].set_ylabel('Score (out of 100)', color='lightgray')
axes[1, 1].tick_params(axis='x', rotation=45)
axes[1, 1].grid(True, linestyle='--', alpha=0.5, axis='y')


plt.tight_layout() # Adjust layout to prevent overlapping
plt.show()

display(HTML("<h3 style='margin-top: 40px;'>The End... Or Is It?</h3><p style='color: #a0a0a0;'>As our tale concludes, we leave Nevermore with a clearer understanding of its haunting inhabitants. But remember, shadows always linger...</p>"))

print("\n--- All Dashboard Visuals Generated ---")
```
